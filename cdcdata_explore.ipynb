```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

```

```python
# Data file
cdcdata_path = "data/Heart_Disease_Mortality_Data_Among_US_Adults__35___by_State_Territory_and_County___2016-2018.csv"

# Read heart disease data
cdcdata_df = pd.read_csv(cdcdata_path)

cdcdata_df
```

```python
cdcdata_rename = cdcdata_df.rename(columns={'Year': 'Year', 'LocationAbbr':'LocationAbbr', 'LocationDesc': 'LocationDesc', 'GeographicLevel':'GeographicLevel', 'DataSource':'DataSource',
       'Class': 'Class', 'Topic': 'Topic', 'Data_Value': "Mortality_Rate", 'Data_Value_Unit': "Morality_Rate_Unit", 'Data_Value_Type': 'Morality_Rate_Type',
       'Data_Value_Footnote_Symbol':'Data_Value_Footnote_Symbol' , 'Data_Value_Footnote':'Mortality_Rate_Footnote',
       'StratificationCategory1':'StratificationCategory1' , 'Stratification1': 'Gender', 'StratificationCategory2':'StratificationCategory2',
       'Stratification2': 'Race', 'TopicID': 'TopicID', 'LocationID':'LocationID', 'Y_lat':'Y_Lat', 'X_lon': "X_Lon"})
```

```python
cdcdata_rename.info()
```

```python
cdcdata_rename.columns
```

```python
cdc_final_cols = ['Year', 'LocationAbbr', 'LocationDesc', 'GeographicLevel', 'DataSource',
       'Class', 'Topic', 'Mortality_Rate', 'Gender','Race', 'TopicID', 'LocationID', 'Y_Lat',
       'X_Lon']
```

```python
cdc_reduced = cdcdata_rename[cdc_final_cols]
```

```python
cdc_reduced.info()
```

```python
cdc_reduced["GeographicLevel"].value_counts()
```

```python
cdc_reduced_states = cdc_reduced[(cdc_reduced.GeographicLevel == 'State')]
cdc_reduced_states["GeographicLevel"].value_counts()
```

```python
cdc_reduced_states.info()
```

```python
cdc_reduced_nation = cdc_reduced[(cdc_reduced.GeographicLevel == 'Nation')]
cdc_reduced_nation.info()
```

```python
cdc_reduced_county = cdc_reduced[(cdc_reduced.GeographicLevel == 'County')]
cdc_reduced_county.info()
```
